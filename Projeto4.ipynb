##Analisando os Dados da ANP

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
df_junho = pd.read_csv('https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsan/2021/2021-06-gasolina-etanol.csv', sep=';')
```

```python
df_julho = pd.read_csv('https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsan/2021/2021-07-gasolina-etanol.csv', sep=';')
```

```python
dataframe = pd.concat([df_junho, df_julho], ignore_index =True)
dataframe
```

##Limpeza dos dados.

```python
df_copia = dataframe.copy()
df_copia
```

```python
df_copia['Valor de Venda'] = df_copia['Valor de Venda'].astype(str)
df_copia['Valor de Venda'] = df_copia['Valor de Venda'].str.replace(',','.')
df_copia['Valor de Venda']
```

```python
df_copia['Valor de Venda'] = df_copia['Valor de Venda'].astype(float)
df_copia['Valor de Venda']
```

```python
df_copia['Valor de Venda'] = df_copia['Valor de Venda'].apply(lambda x: '%.2f' % x)
df_copia
```

```python
df_copia.drop('Valor de Compra', axis='columns', inplace=True)
df_copia.drop('Complemento', axis='columns', inplace=True)
```

```python
df_copia
```

```python
df_copia['Número Rua'] = df_copia['Número Rua'].astype(str)
df_copia['Número Rua'] = df_copia['Número Rua'].str.replace(',', '')
df_copia['Número Rua']
```

```python
df_copia = df_copia.rename(columns={'Região - Sigla':'Região_Sigla', 'Estado - Sigla':'Estado_Sigla', 'CNPJ da Revenda':'CNPJ_Revenda', 'Nome da Rua':'Nome_Rua', 'Número Rua':'Número_Rua', 'Data da Coleta':'Data_Coleta', 'Valor de Venda':'Valor_Venda', 'Unidade de Medida':'Unidade_de_Medida'})
df_copia
```

```python
df_copia['Valor_Venda'] = df_copia['Valor_Venda'].astype(float)
df_copia['Valor_Venda']
```

###Respondendo As Analises 

Qual o preço médio da gasolina e do etanol nesses dois meses?

```python
valor_medio_produto_df = df_copia.groupby('Produto')['Valor_Venda'].mean().map('R$ {:.2f}'.format)
valor_medio_produto_df[:2]
media_dois_meses = pd.DataFrame(valor_medio_produto_df)
media_dois_meses.rename
media_dois_meses.T
```

Qual o preço médio da gasolina e do etanol por estado?

```python
preço_medio = df_copia.groupby(['Produto', 'Estado_Sigla']).mean()
preço_medio['Valor_Venda'] = preço_medio['Valor_Venda'].map('R$ {:.2f}'.format)
preço_medio.unstack()[:2].T
```

Qual a região que possui o maior valor médio da gasolina?

```python
gasolina_max_regiao = df_copia[df_copia['Produto']=='GASOLINA'].groupby(['Região_Sigla', 'Produto']).agg( {"Valor_Venda":"mean"}).round(2).sort_values('Valor_Venda', ascending=False).head(1)
gasolina_max_regiao
```

Qual a região que possui o menor valor médio do etanol?

```python
etanol_min_regiao = df_copia[df_copia['Produto']=='ETANOL'].groupby(['Região_Sigla', 'Produto']).agg( {"Valor_Venda":"mean"}).round(2).sort_values('Valor_Venda').head(1)
etanol_min_regiao
```
